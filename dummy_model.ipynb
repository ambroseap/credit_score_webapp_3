```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option( "display.max_columns", None)

df  = pd.read_csv("creditscore_dummy_dataset - Sheet1.csv") 
```

```python

df['MONTHLY_INCOME_TOTAL'] = df['MONTHLY_INCOME_TOTAL'].fillna(df['MONTHLY_INCOME_TOTAL'].mean())
df['EDUCATIONAL_QUALIFICATION'] = df['EDUCATIONAL_QUALIFICATION'].fillna(df['EDUCATIONAL_QUALIFICATION'].mode()[0])
df['MEANS_OF_IDENTIFICATION'] = df['MEANS_OF_IDENTIFICATION'].fillna(df['MEANS_OF_IDENTIFICATION'].mode()[0])
```

```python
## checking if any categorical Features are there in the Dataset
categorical_data = df.select_dtypes(exclude=[np.number])
print ("There are {} categorical Columns in Dataset".format(categorical_data.shape[1]))
```

```python
d1 = pd.get_dummies(df['GENDER'], drop_first= True)

d2 = pd.get_dummies(df['MARRITAL_STATUS'], drop_first= True)

d3 = pd.get_dummies(df['MEANS_OF_IDENTIFICATION'], drop_first= True)

d4 = pd.get_dummies(df['EMAIL_ADDRESS'], drop_first= True)

d5 = pd.get_dummies(df['TYPE_OF_LOAN'], drop_first= True)

d6 = pd.get_dummies(df['YEARS_IN_PRESENT_EMPLOYMENT'], drop_first= True)


d8 = pd.get_dummies(df['OCCUPATION_TYPE'], drop_first= True)


d10 = pd.get_dummies(df['OWN_A_CAR'], drop_first= True)

d11 = pd.get_dummies(df['OWN_A_HOUSE'], drop_first= True)

d12 = pd.get_dummies(df['EDUCATIONAL_QUALIFICATION'], drop_first= True)

d13 = pd.get_dummies(df['ACCOMODATION_TYPE'], drop_first= True)

d14 = pd.get_dummies(df['APARTMENT_TYPE'], drop_first= True)

d15 = pd.get_dummies(df['LIVING_REGION_CONDITION'], drop_first= True)

d16 = pd.get_dummies(df['DSTV_OR_GOtv_LASTDAY_RECHARGE'], drop_first= True)

d17 = pd.get_dummies(df['NEPA_LASTDAY_PAY'], drop_first= True)

d18 = pd.get_dummies(df['PHONE_LASTDAY_RECHARGE'], drop_first= True)

d19 = pd.get_dummies(df['HOUSE_TYPE'], drop_first= True)

d20 = pd.get_dummies(df['EMPLOYMENT_TYPE'], drop_first= True)

                   












df1 = pd.concat([df, d1, d2, d3, d4, d5,d6,d8,d10,d11,d12,d13,d14,d15,d16,d17,d18,d19,d20], axis = 1)
df=df1

cols = ['GENDER','MARRITAL_STATUS','MEANS_OF_IDENTIFICATION', 'EMAIL_ADDRESS', 'TYPE_OF_LOAN', 'YEARS_IN_PRESENT_EMPLOYMENT',
       'OCCUPATION_TYPE',  'OWN_A_CAR', 'OWN_A_HOUSE', 'EDUCATIONAL_QUALIFICATION',
       'ACCOMODATION_TYPE', 'APARTMENT_TYPE', 'LIVING_REGION_CONDITION', 'DSTV_OR_GOtv_LASTDAY_RECHARGE', 'NEPA_LASTDAY_PAY',
       'PHONE_LASTDAY_RECHARGE', 'EMPLOYMENT_TYPE', 'HOUSE_TYPE' ]
df = df.drop(columns=cols, axis=1)
```

```python
X = df.drop('TARGET', axis = 1)
y = df.TARGET
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
x_train.head()
```

```python
y_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(df)
print(scaled)
```

```python
from sklearn.preprocessing import StandardScaler

# define standard scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(df)
print(scaled)
```

```python
from sklearn.linear_model import LogisticRegression


log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
print("Accuracy is", log_model.score(x_test, y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier(criterion = 'entropy', random_state=100, max_depth=3, min_samples_leaf= 5)

dtree_model.fit(X, y)

predictions = dtree_model.predict(x_test)
print(classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier


r_forset_model = RandomForestClassifier()
r_forset_model.fit(X, y)

predictions = r_forset_model.predict(x_test)
print(classification_report(y_test, predictions))
```

```python
y_pred = r_forset_model.predict(x_test)
```

```python
print("Testing Accuracy :", r_forset_model.score(x_test, y_test))
print("Training Accuracy :", r_forset_model.score(x_train,y_train))
```

```python
stack_model_prob = r_forset_model.predict_proba(x_test)[:,1] # This will give you positive class prediction probabilities  
y_pred = np.where(stack_model_prob > 0.5, 1, 0) # This will threshold the probabilities to give class predictions.
r_forset_model.score(x_test, y_pred)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, stack_model_prob)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```
